```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import csv
import time

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

url = """https://www.melon.com/index.htm"""
driver = webdriver.Chrome(service = Service('driver/chromedriver'))
driver.get(url)
time.sleep(1.5)

xpath = """//*[@id="chk_id01"]"""
driver.find_element(By.XPATH, xpath).click()
time.sleep(1.5)

xpath = """//*[@id="gnb_menu"]/ul[1]/li[1]/a/span[2]"""
driver.find_element(By.XPATH, xpath).click()
time.sleep(1.5)

html = bs(driver.page_source, 'lxml')
time.sleep(1.5)

header = {"User-Agent" : "Mozilla/5.0"}
res = req.get("https://www.melon.com/chart/index.htm", headers =header)
# header 로 페이지 접속 시 접속하는 사람의 정보 전달
# 일반적으로 막혀있어서 헤더 정보를 웹페이지로 변경하여 접근

singer_list = []
song_list = []
#res.text
soup = bs(res.text, 'lxml')

singer = soup.select("span.checkEllipsis")

song = soup.select("div.rank01>span")


for i, k in zip(singer, song):
    singer_list.append(i.text)
    song_list.append(k.text)

print('set')
```
